```python
def showColumnInformation(col):
    global data
    print(f'there are {data[col].isnull().sum()} null Values')
    print(f'there are {data[col].nunique()} unique value')
    print(data[col].unique()[:100])
```

```python
data['approx_cost(for two people)'] = data['approx_cost(for two people)'].astype(str).apply(lambda x : x.replace(',',''))
```

```python
data['approx_cost(for two people)'] = data['approx_cost(for two people)'].astype(float)
```

#### Rest Type

```python
typeCounts = data['rest_type'].value_counts()[:20]
typeCounts
```

```python
print(f"sum of res {sum(data['rest_type'].value_counts()[:20])}")
```

```python
def processRest_type(rest):
    if rest in typeCounts:
        return rest
    else:
        return "other"
```

```python
data['rest_type'] = data['rest_type'].apply(processRest_type)
```
