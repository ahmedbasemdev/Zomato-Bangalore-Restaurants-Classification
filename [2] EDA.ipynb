```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
data = pd.read_csv('processedData.csv')
data.drop('Unnamed: 0',axis=1,inplace=True)
data.head()
```

### sad

```python
def Pie(col,title=""):
    global data
    counts = data[col].value_counts()
    plt.figure(figsize=(12,6))
    plt.pie(counts, labels=counts.index,autopct="%1.1f%%")
    plt.title(title)
    plt.show()
    
def Kde(col,title=""):
    global data
    plt.figure(figsize=(12,6))
    sns.kdeplot(data[col],shade=True)
    plt.title(title)
    plt.show()
def count(col,title=""):
    global data
    plt.figure(figsize=(12,6))
    sns.countplot(data=data , x=col)
    plt.title(title)
    plt.show()
```

```python
Pie('rate',"Rate Precent In Our Data")
```

```python
Kde('votes')
```

```python
data['dish_liked'].value_counts()
```

```python
dishDic = {}
def dishCount(dishes):
    try:
        for dish in dishes.split():
            if dish in dishDic.keys():
                dishDic[dish] +=1
            else:
                dishDic[dish] = 1
    except:
        pass
data['dish_liked'].apply(dishCount)
```
